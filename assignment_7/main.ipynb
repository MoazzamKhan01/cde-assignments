```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Country': ['USA', 'China', 'India', 'Germany', 'Brazil', 'Japan', 'Russia', 'Canada', 'Australia', 'Mexico'],
    'GDP per Capita (USD)': [59850, 10660, 1800, 46100, 10600, 38900, 10750, 41300, 51000, 9500],
    'Life Expectancy (years)': [78.54, 76.91, 69.42, 81.13, 75.88, 84.6, 73.34, 82.2, 82.5, 77.03],
    'Population (millions)': [331, 1393, 1380, 83, 213, 126, 145, 38, 26, 128],
    'Color Grading': [59850, 10660, 1800, 46100, 10600, 38900, 10750, 41300, 51000, 9500]  # Using GDP for color
}

df = pd.DataFrame(data)
```

```python
df 
```

```python
gdp_cap = df['GDP per Capita (USD)']
life_exp = df['Life Expectancy (years)']

pop = df['Population (millions)']

col = ['red', 'green', 'blue', 'blue', 'yellow', 'black', 'green', 'red', 'red', 'green']
plt.scatter(gdp_cap, life_exp, s=pop, c=col)
plt.xscale('log')

plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')

tick_val = [1000, 2000, 3000, 5000, 7000, 10000, 15000, 30000, 50000, 70000, 100000]
tick_lab = ['1k', '2k',  '3k', '5k', '7k', '10k', '15k', '30k', '50k', '70k', '100k']


plt.xticks(tick_val, tick_lab)


plt.text(1750, 70.7, 'India')
plt.text(11000, 78.1, 'China')

plt.grid(True)


plt.show()
```

## Findings: 
- The graph compares GDP per capita (in USD) on the x-axis and Life Expectancy (in years) on the y-axis for different countries.
- Bubble size represents the population of each country — larger bubbles mean higher populations.
- India (blue bubble) has a low GDP per capita (~$2,000) and low life expectancy (~69 years) but a very large population.
- China (green bubble) shows a moderate GDP per capita (~$10,000) and higher life expectancy (~77 years), also with a large population.
- Countries with higher GDP per capita (above $30,000–$70,000) — likely developed nations — show higher life expectancy (above 80 years).
- Developed countries (small bubbles on the right) have smaller populations but longer lifespans.
- There’s a positive correlation between GDP per capita and life expectancy — as wealth increases, people tend to live longer.
- The chart effectively illustrates global inequality in both income and health outcomes across countries in 2007.
