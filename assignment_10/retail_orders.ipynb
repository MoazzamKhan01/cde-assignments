```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("ankitbansal06/retail-orders")

print("Path to dataset files:", path)
```

```python
import pandas as pd

df = pd.read_csv(f"{path}/orders.csv")
df.head()
```

```python
df = pd.read_csv(f"{path}/orders.csv", na_values= ['Not Available', 'unknown'])
df.head()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.shape
```

```python
import numpy as np

df['ship_mode'] = df['ship_mode'].replace(['nan'], np.nan)
```

```python
df.info()
```

```python
df['ship_mode'].fillna(df['ship_mode'].mode()[0], inplace= True)
```

```python
df.info()
```

```python
df['discount']=df['list_price']*df['discount_percent']*.01
```

```python
df['sale_price']= df['list_price']-df['discount']
```

```python
df['profit']=df['sale_price']-df['cost_price']
```

```python
df['order_date']=pd.to_datetime(df['order_date'],format="%Y-%m-%d")
```

```python
df.info()
```

```python
df.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
```

```python
import pyodbc

conn = pyodbc.connect(
    "Driver={ODBC Driver 17 for SQL Server};"
    "Server=tcp:localhost,1433;"
    "Database=master;"
    "Trusted_Connection=yes;"
    "TrustServerCertificate=yes;"
)

print("CONNECTED SUCCESSFULLY")
```

```python
cursor = conn.cursor()

columns = "order_id, order_date, ship_mode, segment, country, city, state, postal_code, region, category, sub_category, product_id, quantity, discount, sale_price, profit"
placeholders = ", ".join(["?"] * 16)
insert_query = f"INSERT INTO orders ({columns}) VALUES ({placeholders})"
```

```python
data_to_insert = [tuple(x) for x in df.values]
```

```python
cursor.fast_executemany = True # Critical for performance
cursor.executemany(insert_query, data_to_insert)
```

```python
conn.commit()
cursor.close()
```
